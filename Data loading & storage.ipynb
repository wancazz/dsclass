```python
#1. I want to see the population estimate from 2010 to 2015 from Washtenaw County
#2. Average population estimate 2015 by STNAME
#3. Count how many counties, CTYNAME
#4. Create a pivoit table showing average BIRTHS2015 for each region
```

```python
#1. I want to see the population estimate from 2010 to 2015 from Washtenaw County
reducedf[reducedf['CTYNAME']=='Washtenaw County'].iloc[0,9:15]
```

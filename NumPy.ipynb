## view and copy

```python
# original array 
arr=np.arange(5)
print(arr)

# the view array
view_arr=arr[1:4]       #basic slicing 
print(view_arr)
view_arr[0]=999        # replace index 0 with number 999
```

```python
print("Array after tempering :")
print("View array :")
print(arr)
print("Original Array")
print(view_arr)    

# check output
# the original also change 
# how?
```

```python
# to avoid tempering original data :

# original array 
arr=np.arange(5)
print(arr)

# the view array (after tempering)
view_arr=arr[1:4].copy()       #basic slicing 
print(view_arr)
view_arr[0]=999        # replace index 0 with number 999

print("Array after tempering :")
print("View array :")
print(arr)
print("Original Array")
print(view_arr)  
```

##### other example with fancy indexing :

```python
# if use fancy slicing, no need to use .copy()

arr=np.arange(5)
print(arr)

# the view array (after tempering)
view_arr=arr[[2,3,4]]          #fancy indexing
print(view_arr)

view_arr[0]=999                #replace index 0 with number 999

print("\nArray after tempering :")
print("View array :")
print(arr)
print("Original Array")
print(view_arr)
```

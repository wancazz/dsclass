```python
data={
    'City':['KL','Selangor','Ipoh','Kuching'],
    'Category':['A','A','B','B'],
    'Sales':[2000,3000,4000,5000]
}
df=pd.DataFrame(data)
df
```

```python
pd.pivot_table(df,values='Sales',columns='City',index='Category')
```

```python
pd.pivot_table(df,values='Sales',columns='City',index='Category',aggfunc='sum')
```

```python
pd.crosstab(df,index='City',columns='Category')
```

```python
pd.crosstab(df['City'],df['Category'])
```

Dictionary

```python
alien_0 = {'color':'green','point':5}
print(alien_0)
```

```python
print(alien_0['color'])
print(alien_0['point'])
```

```python
alien_0['x_position'] = 0
alien_0['y_position'] = 25
print(alien_0)
```

```python
alien_0['color'] = 'yellow'
print(alien_0)
print("The alien is now " + alien_0['color'] + ".")
```

```python
del alien_0['point']
print(alien_0)
```

```python
motorcycles = ['honda','yamaha','suzuki']
```

```python
for motor in motorcycles:
    print('I like ' + motor)
```

```python
customer = {'Name':'John','Status': 'Married', 'Hometown': 'Kuala Lumpur'}

for key, value in customer.items():
    print("\nKey: " + key)
    print("Value: " + value)
```

```python
student = {'Syazwan': 'A123', 'Azman': 'B124', 'Ali':'C125'}

for name, ID in student.items():
    print(name + " is a UKM student with ID number " + ID)
```

```python
ages = {'Ali': 6, 'Wong': 8, 'Ren': 4} #age number is integer here

for name, age in ages.items():
        print(name + " is " + str(age) + " years old.")
```

```python
print("\nThe Following information are required: ")

for info in customer.keys():
    print(info.title())
```

```python
print("\nThe Following information are required: ")

for info in sorted(customer.keys()):
    print(info.title() + " is compulsory. ")
```

```python
print("The details of a customer are listed below:\n")

for info in customer.values():
    print(info.title())
```
